## Code-Painting

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the size of the canvas
canvas_size = 10

# Set the resolution of the painting
resolution = 500

# Create a meshgrid
x = np.linspace(-canvas_size, canvas_size, resolution)
y = np.linspace(-canvas_size, canvas_size, resolution)
X, Y = np.meshgrid(x, y)

# Generate a painting function
Z = np.sin(np.sqrt(X**2 + Y**2)) * np.cos(0.1 * (X**2 + Y**2))

# Create the painting plot
plt.figure(figsize=(8, 8))
plt.imshow(Z, cmap='hot', extent=[-canvas_size, canvas_size, -canvas_size, canvas_size])
plt.colorbar()
plt.title('Painting')
plt.axis('off')

# Save the painting as an image
plt.savefig('painting.png', bbox_inches='tight', pad_inches=0)

# Display the painting
plt.show()
```
